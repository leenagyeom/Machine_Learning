```python
rng = np.random.RandomState() # RandomState seed     
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=color, s=sizes, alpha=0.5, cmap='magma')
plt.colorbar()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
```

```python
plt.scatter(features[0], features[1], alpha=0.2, s=features[3]*100, cmap='plasma', c=iris.target)

plt.xlabel(iris.feature_names[0])
plt.xlabel(iris.feature_names[1])

# iris target    .
```

```python
# errorbar
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.c')
```

```python
plt.style.use('seaborn-whitegrid')
plt.errorbar(x, y, yerr=dy, fmt='o', color='k', ecolor='lightgray', elinewidth=3)
```
