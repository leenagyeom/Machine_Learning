```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def linear_function(x) :
    a = 0.5 # 기울기
    b = 2   # 절편
    
    return a * x + b
```

```python
print(linear_function(5))
```

한개의 스칼라 값이 아니라 벡터(특정한 길이의 배열)를 준다면 업데이트를 해준다.

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)
y
```

```python
# 선형회귀 그래프를 볼 수 있다
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 이차함수

def quadratic_fuction(x) :
    a = 1
    b = -2
    c = -2
    
    return a * x**2 + b * x + c
```

```python
y = quadratic_fuction(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic function')
```

```python
# 3차함수

def cubic_function(x) :
    a = 4
    b = 9
    c = -1
    d = -8
    
    return a * x**3 + b * x**2 + c * x + d
```

```python
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic function')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def AND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.4, 0.4])
  bias = -0.6

  value = np.sum(input * weights) + bias
  #np.sum이 필요한 이유? 뉴런의 수학적 표현에서 시그마 역할을 하기 위해.

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
def OR(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.4, 0.4])
  bias = -0.3

  value = np.sum(input * weights) + bias
  #np.sum이 필요한 이유? 뉴런의 수학적 표현에서 시그마 역할을 하기 위해.

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.3

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='black',marker='^',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='orange',marker='o',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
def NAND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([-0.6,-0.6])
  bias = -0.7

  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

전체적인 구성을 보는 게 중요하다.<br/>

1. 신경망 자체를 만든다.
2. 망 안에 몇 개의 레이어를 쓸건지 고민한다.
3. 각 레이어 안에 뉴런을 몇개 넣을 건지 고민한다.
4. 실제로 만들고, 연결하는 것은 keras에서 알아서 한다.

```python
from keras import models # 모델과 관련된 내용이 들어가있다.
from keras import layers

# 순차적인 망은 Sequential 로 만들 수 있다.
network = models.Sequential()

# 신경망을 만들었으니까 레이어를 추가할 것이다.
# add로 가장 기본적인 Dense 레이어를 추가한다.
# 뉴런은 512개 박는다.
# 활성함수인 activation, 입력으로 들어오는 모양을 28 * 28로 만들었다.
# ,는 한 차원이 더 있다는 걸 나타내는 것이다.
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))

# 출력이 나오는 레이어를 추가해본다.
# Dense 레이어를 추가하고 뉴런은 10개, activation은 softmax를 쓴다.
network.add(layers.Dense(10, activation='softmax'))
```
